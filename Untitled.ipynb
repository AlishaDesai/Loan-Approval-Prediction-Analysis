```python
def replacer(df):
    import pandas as pd
    Q = pd.DataFrame(df.isna().sum())
    Q.columns=["CT"]
    w = list(Q[Q.CT>0].index)
    cat = []
    con = []
    for i in w:
        if(df[i].dtypes=="object"):
            cat.append(i)
        else:
            con.append(i)
    for i in con:
        replacer = df[i].mean()
        df[i] = df[i].fillna(replacer)
    for i in cat:
        replacer = pd.DataFrame(df[i].value_counts()).index[0]
        df[i] = df[i].fillna(replacer)
    return df
```

```python
train = replacer(training_file)
```
